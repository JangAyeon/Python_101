*  read<br>파일의 내용을 하나의 문자열로 읽기<br>**변수.read()**<br>문자 몇 개를 읽을지 정확하게 명시하고자 할 때 사용<br>**변수.read(읽어 올 문자 개수)**<br> 




````
with open("example.txt","r") as file:
  five_chars=file.read(5) #다섯 글자를 읽어옴
  the_rest=file.read() #읽어온 그 뒤부터 바로 통채로 읽어옴

print(five_chars)
print(the_rest)
````

*  for line in file<br>파일의 처음부터 끝까지 모든 줄을 순회<br>반복이 될 때마다 파일의 한 줄을 읽어 온 후, 파일의 커서는 다음 줄 맨 앞으로 이동<br>


````
with open("example.txt",'r') as file:
  for line in file:
    print(line)
    print(len(line))
````

*   file.tell()<br>현재 파일의 위치를 반환<br><br>

````
import tkinter
window=tkinter.Tk()

label=tkinter.Label(window,text="GUI programming")
#Label(레이블이 표시될 컨테이너, 텍스트 이용해 출력될 메시지)
label.pack()
#호출 -> 화면에 레이블 나옴

label.config(text="Hello,Python")
#config 이용해서 레이블...윈도우에 출력할 텍스트를 업데이트

window.mainloop()
````

**사용자로부터 입력 받기**<br>

- Entry<br>
 사용자가 텍스트를 입력할 수 있는 위젯<br>
한 줄짜리 텍스트만 허용<br>
가변변수와 위젯을 연결하면 사용자가 입력할 때마다 위젯의 값 변경 가능






````
import tkinter
window=tkinter.Tk()

#문자열 가변변수 생성
var=tkinter.StringVar()

#변수 var를 레이블과 엔트에 연결
# -> 엔트리로 입력한 값에 따라서 다른 연산 수행 가능
# -> 쓸 때마다 위에 있는 레이블 값이 바뀜
# 엔트리에 입력된 값들이 변수로 저장되고 레이블로 그 값을 출력
label=tkinter.Label(window,textvariable=var)
label.pack()
entry=tkinter.Entry(window,textvariable=var)
entry.pack()

window.mainloop()
````

- 이벤트 핸들링<br>버튼을 클릭하거나 메뉴를 클릭할 때, 특정한 동작을 수행하도록 한 것<br>이벤트 처리를 위한 속성 'command=' 옵션 이용해서 처리



````
import tkinter
from tkinter import messagebox #알림창을 표시하기 위함

window=tkinter.Tk()


#버튼이 클릭되었을 때 실행될 이벤트 처리 함수
def buttonClick():
    name=entry.get() #Entry로 부터 문자열을 읽거와 name에 저장
    messagebox.showinfo(name,"Button Click")


label=tkinter.Label(window,text="이름")
label.pack()
entry=tkinter.Entry(window)
entry.pack()

#버튼을 클릭하면 buttonClick 함수를 호출하도록 설정
button=tkinter.Button(window,text="Button",command=buttonClick)
button.pack()

window.mainloop()
````

- 이벤트 바인딩<br>이벤트와 이벤트 핸들러를 연결해 줌<br>**Blind(이벤트명, 이벤트 핸들러)**

**이벤트명 / 설명**<br>image.png




````
import tkinter
window=tkinter.Tk()


#마우스 왼쪽 버튼이 클릭되었을 때 실핼될 이벤트 처리 함수
#이벤트가 발생된 위치에 대해 x좌표, y좌표 (왼쪽 상단-(0,0), 오른쪽 하단-(300,300))
def mouseClick(event):
    print("마우스 클릭 위치: ",event.x,event.y)

frame=tkinter.Frame(window,width=300,height=300)


#이벤트 바인
#마우스 왼쪽 버튼이 클릭이 되면 mouseClick이라는 함수 실
frame.bind("<Button-1>",mouseClick)

frame.pack()
window.mainloop()
````

**위젯 배치**<br>


*   tkinter에서 기본적으로 사용하는 레이아웃은 위에서 아래 순서대로 배치

*   단순 배치 : pack()<br>pack(side=값)<br>pack(side="left") / pack(side="right")
*   세밀한 배치 제어 : grid 이용<br>gird : 윈도우와 프레임을 행과 열의 그리드 처리<br><br>grid 매개변수<br>row : 위젯을 위치 시킬 행, 0 ~<br>column : 위젯을 위치 시킬 열, 0 ~<br>rowspan : 위젯이 차지하는 행 갯수, default=1<br>columnspan : 위젯이 차지하는 열 갯수, default=1
