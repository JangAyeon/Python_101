#1번 : 원의 면적과 둘레 구하기
image.png

```
PI=3.14159

def circle_area(r): 
  #복사하여 새롭게 생성된 값을 매개변수 r에 할당, 지역변수 r이 그 값을 가르키게 됨
  #같은 값이라고 하더라도 서로 다른 공간에 저장됨
  return PI*r*r

def circle_circumference(r):
  return 2*PI*r

r=float(input("원의 반지름을 입력하시오: ")) #(1)shell에서 생성된 r값을 함수에 복사해 전달
print("원의 면적: ",circle_area(r))
print("원의 둘레: ",circle_circumference(r))
```

#2번 : 평균 구하기

image.png

```
def average(score1,score2,score3,score4):
  #점수가 0과 100 사이의 수가 아닌 경우 함수 종료
  if score1>100 or score1<0:
    print("invalid input")
    return
  elif score2>100 or score2<0:
    print("invalid input")
    return
  elif score3>100 or score3<0:
    print("invalid input")
    return
  elif score4>100 or score4<0:
    print("invalid input")
    return
  
  #상위 3개의 점수 평균 구하기 위해 점수값 내림차순으로 정렬
  score_list=[score1,score2,score3,score4]
  score_list.sort()
  score_list.reverse()

  #상위 3개의 점수 총합 계산
  sum=0
  for i in range(0,3):
    sum+=score_list[i]

  #상위 3개의 점수 총합을 이용한 평균 계산
  result=sum/3
  return result

average(85,92,80,90)
```

```
def average(score1,score2,score3,score4):
  lst=[score1,score2,score3,score4]
  for i in range(len(lst)):
    if i>100 or i<0:
      print("invalid input")
      return
  score=[]
  for i in range(3):
    score.append(lst.pop(lst.index(max(lst))))
  return sum(score)/len(score)

average(85,92,80,90)
```

```
contact_list = [] # 연락처를 저장할 리스트 (전역변수)

def display_list(): #함수 안에서 함수 호출할 때는 해당 함수 이미 선언되어 있어야함
  for i in range(len(contact_list)):
    print("이름: ",contact_list[i][0],"연락처: ",contact_list[i][1])

def add_list():
  name = input('추가할 이름을 입력하세요: ')
  call = input('추가할 전화번호를 입력하세요: ')
  contact_list.append([name, call])

def contact() : # 연락처 관리 함수
# 사용자(키보드)로부터 연락처 관리를 수행할 메뉴를 입력 받음
  while True:

    print('''메뉴\t설명
1\t전체 목록 출력
2\t연락처 추가
3\t연락처 삭제
4\t종료''')
    menu = int(input('연락처 관리 프로그램 메뉴를 선택하세요'))
    if menu == 1 : # contact_list에 있는 모든 목록 출력
      display_list()
    elif menu == 2 : # 새로운 연락처 추가
      add_list()
    elif menu == 3 : # 기존 연락처 삭제
      name = input('삭제할 이름을 입력하세요: ')
      call = input('삭제할 전화번호를 입력하세요: ')
      contact_list.remove([name, call])
    elif menu==4:
      break
    else :
      print('잘못 입력하셨습니다.')

contact()

```
