#1번 : 키 입력 이벤트<br>image.png

```
import tkinter
window=tkinter.Tk()

label1=tkinter.Label(window,text="Press any key") #레이블 한 개
label1.pack()

var=tkinter.StringVar() #빈 문자의 두번째 레이블
var.set(" ")
label2=tkinter.Label(window,textvariable=var)
label2.pack()

def ehandler(event):
  var.set("key: %s"%event.char)

window.bind("<Key>",ehandler)
window.mainloop()
```

#3번 : 계산기<br>image.png

```
import math
import tkinter

window=tkinter.Tk()

ep=tkinter.StringVar()
entry=tkinter.Entry(window,textvariable=ep) #수식 입력받을 엔트리 생성
entry.pack(side="left") #엔트리 배치

label=tkinter.Label(window,text=" ")
label.pack(side="right")


def onClick(event):
    label.config(text=str(eval(ep.get())))
    
button=tkinter.Button(window, text="계산",command=onClick)
button.pack()

window.mainloop()
```
