```

import random

def dice():
  return random.randint(1,6) #1이상 6이하의 랜덤한 정수 

while True:
  print("주사위의 값:"+str(dice())) #문자열 연산 
  cont=input("계속하시겠습니까?(Y/N): ")

  if (cont.upper()=="Y") or (cont.upper()=="YES"):
    continue
  else:
    break
```

```
# temperature.py
def convert_to_cel(fah) :
    return (fah - 32.0) * 5.0 / 9.0
temp = float(input("화씨 온도를 입력하세요: "))
cel = str(convert_to_cel(temp))
print("섭씨 온도는 " + cel + "입니다")
```

```
#baking.py
import temperature

temp=float(input("조리 온도를 화씨로 입력하시오: "))
cel=temperature.convert_to_cel(temp) #화씨 -> 섭씨
print("오븐을 섭씨"+str(cel)+"도로 예열하시오.")
```
