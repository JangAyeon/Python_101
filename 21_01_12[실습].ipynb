#1번 : 집합 연산<br>
image.png

```
A={'a','n','e','t','x','y','n','w'}
B={'p','y','t','h','o','n'}

print("Set 함수")
print("합집합",A.union(B))
print("교집합",A.intersection(B))
print("차집합",A.difference(B))
print("대칭차집합",A.symmetric_difference(B))

print("연산자")
print("합집합",A|B)
print("교집합",A&B)
print("차집합",A-B)
print("대칭차집합",A^B)
```

#2번 : 모스 부호 <br>
image.png

```
morse = { 'A':'._', 'B':'_...', 'C':'_._.', 'D':'_..', 'E':'.', 'F':'.._.',
'G':'__.', 'H':'....', 'I':'..', 'J':'.___', 'K':'_._', 'L':'._..',
'M':'__', 'N':'_.', 'O':'___', 'P':'.__.', 'Q':'__._', 'R':'._.',
'S':'...', 'T':'_', 'U':'.._', 'V':'..._', 'W':'.__', 'X':'_.._',
'Y':'_.__', 'Z':'__..' } 


def print_morse(s):
  result=""
  s=s.upper()

  for char in s:
    if char==" ":
      continue
    result+=morse[char]

  return result

print(print_morse("Hello Python"))
print(print_morse("hello python"))
print(print_morse("123")) #해당하는 키 값 없어서 오류 발생
```

```
morse = { 'A':'._', 'B':'_...', 'C':'_._.', 'D':'_..', 'E':'.', 'F':'.._.',
'G':'__.', 'H':'....', 'I':'..', 'J':'.___', 'K':'_._', 'L':'._..',
'M':'__', 'N':'_.', 'O':'___', 'P':'.__.', 'Q':'__._', 'R':'._.',
'S':'...', 'T':'_', 'U':'.._', 'V':'..._', 'W':'.__', 'X':'_.._',
'Y':'_.__', 'Z':'__..' } 


def print_morse(s):
  result=""
  s=s.upper()

  for char in s:
    if char=='':
      continue
    result+=morse.get(char,'') #키 값으로 없는 경우 공백문자 출력

  return result

print(print_morse("123")) #해당하는 키 값 없어서 그냥 공백 출력됨
```

#3번 : 문자열 역순 출력<br>
image.png

```
str1=input("문자열을 입력하시오: ")
tmplist=list(str1)
tmplist.reverse()
result=''
for char in tmplist:
  result+=char

print(result)
```

```
#함수로 만들기

def reverse_txt(str1):

  result=''
  for i in range(len(str1)-1,-1,-1):
    result+=str1[i]
  return result

reverse_txt("NORMAL TEXT")
```

#4번 : 숫자 출력<BR>image.png


```
dicKor = { '일':'one', '이':'two', '삼':'three' ,'사':'four','오':'five','육':'six','칠':'seven','팔':'eight','아홉':'nine','십':'ten'}
dicEng = { 'one':'일', 'two':'이', 'three':'삼','four':'사','five':'오','six':'육','seven':'칠','eight':'팔','nine':'구','ten':'십' }
word = input('검색할 단어를 입력하세요: ')
if word in dicKor.keys() :
  print(dicKor[word])
elif word in dicEng.keys() :  
  print(dicEng[word])
else :
  print('잘못 입력하셨습니다.')

```

```
#함수로 만들기
dicKor = { '일':'one', '이':'two', '삼':'three' ,'사':'four','오':'five','육':'six','칠':'seven','팔':'eight','아홉':'nine','십':'ten'}
dicEng = { 'one':'일', 'two':'이', 'three':'삼','four':'사','five':'오','six':'육','seven':'칠','eight':'팔','nine':'구','ten':'십' }

def search_word(word):
  if word in dicKor.keys():
    return dicKor[word]
  elif word in dicEng.keys():
    return dicEng[word]
  else:
    return "invalid input"

word=input("검색할 단어를 입력하시오: ")
print(search_word(word))
```
